```python
import os
from dotenv import load_dotenv
load_dotenv()

from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.output_parsers import StrOutputParser
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.prompts.chat import SystemMessagePromptTemplate, HumanMessagePromptTemplate

# Initialize the model
chatGroq_model = ChatGroq(model="llama-3.3-70b-versatile")

# Create a generic system + human prompt template
prompt = ChatPromptTemplate.from_messages([
    SystemMessagePromptTemplate.from_template(
        "You are a helpful translator that translates English text to any language requested."
    ),
    HumanMessagePromptTemplate.from_template(
        "Translate the following text to {language}: {text}"
    )
])

parser = StrOutputParser()
chain = chatGroq_model | parser

# Chat loop
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit", "bye"]:
        break

    # Fill the template
    response = prompt.invoke({
        "language": "French",  # you can make this dynamic too
        "text": user_input
    })

    print(f"You: {user_input}")
    print(f"Groq: {response}")
```
