```python
# %pip install langchain_chroma
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import TextLoader
from langchain_chroma import Chroma

loader = TextLoader("textfile.txt")
file = loader.load()

file
######################################
splitter = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=50)

documents = splitter.split_documents(file)

documents
########################################
ollama_embeddings = OllamaEmbeddings(model='nomic-embed-text:latest')



# print(embeddings[0])
# print(embeddings[1])


db = Chroma.from_documents(documents=documents,embedding=ollama_embeddings, persist_directory="./chromadb")
db

query = "what is an essay ?"

res = db.similarity_search(query)
```
